```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
# let's load the file
#df = pd.read_csv('C:/Users/suraj/Downloads/sales_data_sample.csv',sep=",", encoding='Latin-1')
#print(df.head())

print(sns.get_dataset_names())
crash_df = sns.load_dataset('car_crashes')
tips_df = sns.load_dataset('tips')
print(crash_df.head())
```

## Distribution Plot

```python
sns.distplot(crash_df['not_distracted'],kde = False, bins = 25)
```

## Joint Plot

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex') #reg
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol'], label = 'alcohol')
```

## Pair Plots 

```python
sns.pairplot(crash_df)
```

```python
tips_df.head()
```

```python
sns.pairplot(tips_df, hue = 'sex', palette = 'Blues')
```

## Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

## Styling


```python
sns.set_style('white') #darkgrid #whitegrid #dark

plt.figure(figsize=(8,4))

sns.set_context('poster', font_scale=1.4)
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')


```

## Categorical Plot

## Bar Plot

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips_df, estimator = np.median)
```

## Count Plot

```python
sns.countplot(x = 'sex', data = tips_df)
```

## Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')
plt.legend(loc = 0)
```

## Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')
```

## Strip Plot

```python
plt.figure(figsize  = (8,5))

sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True, hue = 'sex')
```

## Swarm Plot

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips_df)
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, color = 'white')
```

## Palattes

```python
plt.figure(figsize=(8,6))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', palette='magma')
plt.legend(loc=0)
```

## Heatmaps

```python
crash_df.head()
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)


crash_df_1 = crash_df.iloc[:, :-1]

crash_df_1.head()

crash_mx = crash_df_1.corr()
crash_mx.head()

sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
```

## Cluster Map

```python
iris = sns.load_dataset('iris')

species = iris.pop('species')
sns.clustermap(iris)
```

## Pairgrid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

## Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df,col = 'time', row = 'smoker', height = 4, aspect = 1, palette='Set1')

tips_fg.map(plt.hist, 'total_bill', bins =8)
```
